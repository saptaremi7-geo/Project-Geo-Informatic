# 🌍 Global Coordinate Transformation Tool - Version 1.0

### 📌 Overview
Program ini dibuat untuk melakukan **transformasi sistem koordinat secara otomatis** dari file Excel yang diupload pengguna.  
Tool ini mendukung transformasi koordinat secara **global** dengan fitur **auto-detect zona UTM**.

---

### 🔄 Supported Transformations
1️⃣ **UTM → Latitude / Longitude (WGS84)**  
2️⃣ **Latitude / Longitude → UTM (Auto Detect Zone & Hemisphere)**  


---

### 📂 Input
File **Excel (.xlsx)** dengan format kolom koordinat:

| X | Y |
|---|---|
| Easting / Longitude | Northing / Latitude |

---

### 📤 Output
Hasil transformasi akan diekspor dalam bentuk **Excel (.xlsx)** dengan sistem koordinat baru.

- PyProj (PROJ Geodetic Library)

```python
# ==========================================================
# UNIVERSAL COORDINATE TRANSFORMER (UTM & LATLON)
# Author : Remilla Sapta Nada
# Engine : PROJ (pyproj)
# Version : 1.0
# ==========================================================

!pip install pyproj pandas openpyxl

import pandas as pd
from pyproj import Transformer
from google.colab import files

print("===================================")
print("UPLOAD FILE (Excel / CSV)")
print("===================================")

uploaded = files.upload()

file_name = list(uploaded.keys())[0]

# ===============================
# READ FILE
# ===============================

if file_name.endswith(".xlsx"):
    df = pd.read_excel(file_name)

elif file_name.endswith(".csv"):
    df = pd.read_csv(file_name)

else:
    raise Exception("Format file tidak didukung")

print("\nKolom yang terbaca:")
print(df.columns)

# ===============================
# AUTO DETECT KOLOM
# ===============================

x_candidates = ["X","Easting","x","easting"]
y_candidates = ["Y","Northing","y","northing"]

lon_candidates = ["Lon","Longitude","lon","longitude"]
lat_candidates = ["Lat","Latitude","lat","latitude"]

x_col = None
y_col = None
lon_col = None
lat_col = None

for col in df.columns:

    if col in x_candidates:
        x_col = col

    if col in y_candidates:
        y_col = col

    if col in lon_candidates:
        lon_col = col

    if col in lat_candidates:
        lat_col = col


# ===============================
# MENU TRANSFORMASI
# ===============================

print("\n===================================")
print("PILIH TRANSFORMASI")
print("===================================")

print("1 = UTM -> LatLon")
print("2 = LatLon -> UTM (Auto Detect Zone)")

mode = input("Masukkan pilihan : ")

# =====================================================
# 1 UTM -> LATLON
# =====================================================

if mode == "1":

    if x_col is None or y_col is None:
        raise Exception("Kolom X / Easting dan Y / Northing tidak ditemukan")

    zone = int(input("Masukkan Zona UTM : "))
    hemisphere = input("Hemisphere (N/S) : ")

    if hemisphere.upper()=="S":
        epsg = 32700 + zone
    else:
        epsg = 32600 + zone

    transformer = Transformer.from_crs(f"EPSG:{epsg}", "EPSG:4326", always_xy=True)

    lon, lat = transformer.transform(df[x_col].values, df[y_col].values)

    df["Longitude"] = lon
    df["Latitude"] = lat


# =====================================================
# 2 LATLON -> UTM (AUTO DETECT ZONE)
# =====================================================

elif mode == "2":

    if lon_col is None or lat_col is None:
        raise Exception("Kolom Longitude / Latitude tidak ditemukan")

    lon = df[lon_col].values
    lat = df[lat_col].values

    easting_list = []
    northing_list = []
    zone_list = []

    for i in range(len(lon)):

        zone = int((lon[i] + 180) / 6) + 1

        if lat[i] >= 0:
            epsg = 32600 + zone
            hemi = "N"
        else:
            epsg = 32700 + zone
            hemi = "S"

        transformer = Transformer.from_crs("EPSG:4326", f"EPSG:{epsg}", always_xy=True)

        x,y = transformer.transform(lon[i], lat[i])

        easting_list.append(x)
        northing_list.append(y)
        zone_list.append(str(zone)+hemi)

    df["UTM_Easting"] = easting_list
    df["UTM_Northing"] = northing_list
    df["UTM_Zone"] = zone_list


else:

    raise Exception("Pilihan tidak valid")

# =====================================================
# EXPORT
# =====================================================

output="hasil_transformasi_koordinat.xlsx"

df.to_excel(output,index=False)

print("\n===================================")
print("TRANSFORMASI SELESAI")
print("File akan terdownload")
print("===================================")

files.download(output)
```
